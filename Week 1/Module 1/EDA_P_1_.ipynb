7.Univariate Analysis(Numerical Column Analysis-1)

```python
sns.histplot(data=df,x=df['Age'],bins=30)
plt.xlabel='Age'
plt.ylabel='Count'
plt.title('Distribution of Age')
plt.show()
```

```python
sns.kdeplot(df['Age'])
plt.title('Kde plot of Age')
plt.xlabel('Ages')
plt.ylabel('Probability distribution')
plt.show()
```
