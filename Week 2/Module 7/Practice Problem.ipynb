Use the Pima Indians Diabetes Database dataset. 

Tasks: 

●  Load the dataset using Pandas.

●  Apply: 

  ○  Necessary Imputation ( Must be done with the proper one ) 

●  Create a categorical column:

```python
import pandas as pd
import numpy as np
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#Replace invalid 0 values in columns like Glucose, BloodPressure, and BMI with NaN. 
columns_to_replace=['Glucose', 'BloodPressure', 'BMI']
for column in columns_to_replace:
     df[column]=df[column].replace(0,np.nan)
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer =SimpleImputer(strategy='mean')
df_imputed=imputer.fit_transform(df)
df_imputed=pd.DataFrame(df_imputed,columns=df.columns)
df_imputed.isnull().sum()
```

```python
df_imputed
```

 Create a categorical column: 

 ○  Age_Group 

■   Young (<30) 

■   Adult (30–50)
 
■   Senior (>50) 

```python
def Age_Group(age):
    if age<30:
        return 'Young'
    elif age<50:
        return 'Adult'
    else:
        return 'Senior'
df_imputed['Age_Group']=df_imputed['Age'].apply(Age_Group)
df_imputed.head()
```

```python
#another way
df_imputed['Age_Group']=df_imputed['Age_Group'].apply(lambda x: 'Young' if x=='Young' else ('Adult' if x=='Adult' else 'Senior'))
df_imputed.head()
```

```python
 #Apply Encoding on Age_Group. Choose the proper encoding technique and explain why it is the best one
 #ordinalEncoder
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder(categories=[['Senior','Adult','Young']])
df_imputed['Age_Group_Encoded']=encoder.fit_transform(df_imputed[['Age_Group']])
df_imputed.head()
 
```

```python
#level encoding
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df_imputed['Age_Group_Label_Encoded']=label_encoder.fit_transform(df_imputed[['Age_Group']])
df_imputed.head()
 
```

```python
print(df_imputed[
    ['Age_Group',
     'Age_Group_Encoded',
     'Age_Group_Label_Encoded']
].head())
```

# 6. Diabetes Dataset – Compare Encoding Techniques 

Dataset Link: 

Pima Indians Diabetes Dataset 

Use the Pima Indians Diabetes Database dataset. 

Tasks: 

●  Create a categorical feature from BMI: 

      ○  Underweight 

      ○  Normal 
    
      ○  Overweight 

      ○  Obese 

●  Apply: 

      ○  Label Encoding 

      ○  Ordinal Encoding 

      ○  One Hot Encoding 

●  Write a short conclusion on which encoding is best for this feature. 
 



```python
df=pd.read_csv('diabetes.csv')
df
def find_weight(BMI):
    if BMI <18.5:
        return 'Underweight'
    elif BMI<18.5:
        return 'Normal'
    elif BMI<25:
        return 'Overweight'
    else:
        return 'Obese'

df['health']=df['BMI'].apply(find_weight)
df
```

```python
#  Label Encoding 
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df['health_encoded']=label_encoder.fit_transform(df['health'])
df['health_encoded']


```

```python
#Ordinal Encoding 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
X_train, X_test, y_train, y_test = train_test_split(df.drop('health', axis=1), df['health'], test_size=0.2, random_state=42)
ordinal_encoder = OrdinalEncoder(categories=[['Underweight', 'Normal', 'Overweight','Obese']])

y_train_encoded = ordinal_encoder.fit_transform(y_train.to_frame())

print(pd.DataFrame({'health': y_train, 'health_encoded': y_train_encoded.flatten()}).head())
```

```python
#One Hot Encoding 
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder=OneHotEncoder(sparse_output=False,drop='first')
encoded_health=one_hot_encoder.fit_transform(df[['health']])
encoded_health_df=pd.DataFrame(encoded_health,columns=one_hot_encoder.get_feature_names_out(['health']))
encoded_health_df.head()
```
